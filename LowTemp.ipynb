```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from matplotlib import font_manager as fm, rcParams
```

```python
font_sz = 18

fig, ax = plt.subplots()

fpath = os.path.join(rcParams["datapath"], "fonts/ttf/cmr10.ttf")
prop = fm.FontProperties(fname=fpath,size=font_sz)
fname = os.path.split(fpath)[1]
ax.set_title('This is a special font: {}'.format(fname), fontproperties=prop)
ax.set_xlabel('This is the default font')

plt.show()
```

```python
data0 = np.loadtxt("data/228_liquidN2.txt")
print(data0.shape)
```

```python
def tvo(data,title,ymin=None,ymax=None):
    plt.scatter(data[:,1],data[:,3]*1E6, alpha=0.5)
    plt.title(title,fontproperties=prop)
    plt.xlabel("Temperature(K)",fontproperties=prop)
    plt.ylabel("Resistance(Ohm)",fontproperties=prop)
    axes = plt.gca()
    axes.set_ylim([ymin,ymax])
    fig = plt.gcf()
    fig.set_size_inches(8,6)
    plt.savefig("{}.png".format(title))
    plt.show()
```

```python
def tvv(data,title,ymin=None,ymax=None):
    plt.scatter(data[:,1],data[:,3], alpha=0.5)
    plt.title(title,fontproperties=prop)
    plt.xlabel("Temperature(K)",fontproperties=prop)
    plt.ylabel("Voltage(mV)",fontproperties=prop)
    axes = plt.gca()
    axes.set_ylim([ymin,ymax])
    fig = plt.gcf()
    fig.set_size_inches(8,6)
    plt.savefig("{}.png".format(title))
    plt.show()
```

```python
tvv(data0,"Liquid Nitrogen Cooling (A-B)")
tvo(data0,"Liquid Nitrogen Cooling (A-B)")
```

```python
data2 = np.loadtxt("data/302_heat20v.txt")
print(data2.shape)
```

```python
temp = data2[11:196,1]
```

```python
dtemp = np.array([])
for i in range(temp.shape[0]-3):
    dtemp = np.append(dtemp,temp[i+3]-temp[i])
```

```python
gamma = 0.695E-3
beta = 4.82E-5
def h(w,h):
    plt.title("Liquid Helium Heat Capacity Versus Temperature", fontproperties=prop)
    plt.xlabel("Temperature(K)", fontproperties=prop)
    plt.ylabel(r'C($\frac{J}{mol \cdot K}$)', fontproperties=prop)
    plt.setp(ax.get_xticklabels(),fontproperties=prop)
    plt.scatter(temp[:dtemp.shape[0]],0.5/dtemp-gamma*dtemp-beta*dtemp**3,alpha=0.5)
    plt.axvline(x=2.45, color='r', linestyle='--')
    fig = plt.gcf()
    fig.set_size_inches(w,h)
    plt.savefig("h.png")
    plt.show()
```

```python
h(8,6)
```
